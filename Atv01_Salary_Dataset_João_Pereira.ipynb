```python
from google.colab import files
import pandas as pd
import io
```

```python
uploaded_file = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded_file['Salary_dataset.csv']))
display(df.head())
```

```python
# checando a distribuio que existe entre as variveis por meio do scatter plot:

import matplotlib.pyplot as plt

plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# importando todas as bibliotecas para aplicar o modelo de regresso linear:

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# pegando as linhas e colunas do dataset  (Predio de salrio a partir dos anos de experincia)

# X_trains  a varivel independente anos de experincia (YearsExperience)
# y_trains  a varivel alvo (feature)

X_train = df[['YearsExperience']]
y_train = df['Salary']
```

```python
X_test = df[['YearsExperience']]
y_test = df['Salary']
```

```python
# pegando os testes das features para usar o modelo Regresso linear para validar o resultado das features:

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = model.score(X_test, y_test)
print('R2:', r2)

print(f"Erro Quadrtico Mdio (MSE): {mse:.2f}")
print(f"Coeficiente de Determinao (R): {r2:.2f}")
print(f"Intercepto: {model.intercept_:.2f}")
print(f"Coeficiente angular: {model.coef_[0]:.2f}")
```

# Concluso:

##### Com base no modelo de Regresso Linear aplicado ao dataset, observamos que existe uma relao positiva e linear entre os anos de experincia e o salrio. Isso significa que, conforme o profissional acumula mais anos de experincia, o salrio tende a aumentar proporcionalmente.

O modelo obteve um R prximo de 1 (ou um valor elevado de acordo com sua sada), indicando que uma grande parte da variao do salrio  explicada diretamente pelos anos de experincia. Ou seja, o modelo possui boa capacidade de previso para esse conjunto de dados.

Alm disso:

O coeficiente angular (coef_) mostrou o quanto o salrio aumenta, em mdia, a cada ano adicional de experincia.

O intercepto indica o valor estimado do salrio quando a experincia  zero.

O Erro Quadrtico Mdio (MSE) apresentou um valor baixo, o que refora que as previses do modelo esto prximas dos valores reais.

Portanto, o modelo se mostrou coerente, estatisticamente consistente e adequado para prever salrios com base na experincia, sendo um exemplo claro de regresso linear simples e eficaz.
