```python
from google.colab import files
import pandas as pd
import io
```

```python
uploaded_file = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded_file['Salary_dataset.csv']))
display(df.head())
```

```python
# checando a distribuição que existe entre as variáveis por meio do scatter plot:

import matplotlib.pyplot as plt

plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# importando todas as bibliotecas para aplicar o modelo de regressão linear:

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# pegando as linhas e colunas do dataset  (Predição de salário a partir dos anos de experiência)

# X_trains é a variável independente anos de experiência (YearsExperience)
# y_trains é a variável alvo (feature)

X_train = df[['YearsExperience']]
y_train = df['Salary']
```

```python
X_test = df[['YearsExperience']]
y_test = df['Salary']
```

```python
# pegando os testes das features para usar o modelo Regressão linear para validar o resultado das features:

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = model.score(X_test, y_test)
print('R2:', r2)

print(f"Erro Quadrático Médio (MSE): {mse:.2f}")
print(f"Coeficiente de Determinação (R²): {r2:.2f}")
print(f"Intercepto: {model.intercept_:.2f}")
print(f"Coeficiente angular: {model.coef_[0]:.2f}")
```

# Conclusão:

##### Com base no modelo de Regressão Linear aplicado ao dataset, observamos que existe uma relação positiva e linear entre os anos de experiência e o salário. Isso significa que, conforme o profissional acumula mais anos de experiência, o salário tende a aumentar proporcionalmente.

O modelo obteve um R² próximo de 1 (ou um valor elevado de acordo com sua saída), indicando que uma grande parte da variação do salário é explicada diretamente pelos anos de experiência. Ou seja, o modelo possui boa capacidade de previsão para esse conjunto de dados.

Além disso:

O coeficiente angular (coef_) mostrou o quanto o salário aumenta, em média, a cada ano adicional de experiência.

O intercepto indica o valor estimado do salário quando a experiência é zero.

O Erro Quadrático Médio (MSE) apresentou um valor baixo, o que reforça que as previsões do modelo estão próximas dos valores reais.

Portanto, o modelo se mostrou coerente, estatisticamente consistente e adequado para prever salários com base na experiência, sendo um exemplo claro de regressão linear simples e eficaz.
